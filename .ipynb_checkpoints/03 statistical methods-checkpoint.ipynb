Taking a rolling mean is a simple way to remove seasonality from our data and get a trend
This can be useful as an engineered feature. If you start from 1950 the trend isn't linear. There are many dangerous assumptions baked in here.

```python
tavg["1960-01-01":].rolling(3650).mean().plot()
```

```python
rolling_mean =tavg.rolling(366).mean()
rolling_mean.plot()
```

```python
rolling_mean.loc["1951-12-25":"1952-01-05"]
# notice exactly one year Nan, our window is backwards looking.
```

By taking the difference between time t0 and t1 we remove the current level and seasonality from our data to some extent. This allows us to observe the amount of change or volatility in our data from one day to the next.

```python
tavg_diff = tavg.diff()
tavg_diff["2000-01-01":].plot()
# this is probably useless, but i left it in anyway.
#Just because you can plot something doesn't mean it's useful
```

```python
rolling_mean_diff = rolling_mean.diff()

rolling_mean_diff["2000-01-01":].plot()
```

Taking the difference between the rolling mean of one day and the next day gives us a (smoothed) plot of the change in temperature from one day to the next. 

```python
dftest = adfuller(rolling_mean_diff.dropna(), autolag = 'AIC')
p_value = dftest[1]


print("p value:",p_value, "statistic",dftest[0]) # low p value indicates stationary

# again, I'm not sure this is actually useful
```

```python
weekly = tavg.resample("W").mean()
weekly["2010-01-01":].plot()
```

```python
# we use all the data up to 2015 for our model
xs = ExponentialSmoothing(weekly[:"2015-01-01"],trend='add',seasonal='add',seasonal_periods=52, freq="W")
fitted_model = xs.fit()
```

```python
# now we forecast multiple years ahead 
start_date = len(weekly[:"2015-01-01"])
end_date = len(weekly[:"2021-01-01"])


actual = weekly["2015-01-01":"2021-01-01"]

#the benchmark will be the last few years measurements
# so we can compare our forecast to the last few years to see if there will be any improvement
benchmark = pd.DataFrame(index=actual.index)
benchmark["benchmark"] = weekly["2009-01-01":"2015-01-01"].values

#fit the model
prediction = fitted_model.predict(start=start_date, end=end_date )


comparison = pd.concat([actual, prediction, benchmark], axis=1)
comparison.columns = ["actual", "predicted","benchmark"]
comparison
```

```python
comparison["2018-01-01":].plot()
```

```python
comparison.fillna(method="ffill", inplace=True)
mse_value = mse(comparison["actual"], comparison["predicted"])
mae_value = mae(comparison["actual"], comparison["predicted"])
r2 = r2_score(comparison["actual"], comparison["predicted"])
mse_value, mae_value, r2
```

```python
mse_value = mse(comparison["benchmark"], comparison["predicted"])
mae_value = mae(comparison["benchmark"], comparison["predicted"])
r2 = r2_score(comparison["benchmark"], comparison["predicted"])
mse_value, mae_value, r2
```

```python
model=sm.tsa.statespace.SARIMAX(weekly["1980-01-01":"2018-01-01"],order=(1, 1, 1),seasonal_order=(1,1,1,52))
```

```python
results=model.fit()
```

```python
weekly['forecast']= results.predict(start="2018-01-01", end="2022-12-31")
weekly["2018-01-01":].plot()
```

```python
x = weekly["measurement"].loc["2018-01-01":]
y = weekly["forecast"].loc["2018-01-01":]
mae(x,y)
```

```python
r2 = r2_score(y, x)
r2
```
