```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.seasonal import STL
```

```python
df1 = pd.read_csv("new_england_tmax.csv")
df2 = pd.read_csv("new_england_tmin.csv")

df1.index = pd.to_datetime(df1['date'])
df2.index = pd.to_datetime(df2['date'])

df1 = df1.loc["1951-01-01":"1970-01-01"]
df2 = df2.loc["1951-01-01":"1970-01-01"]

x = df1.index

# # Variables on y axis
y = df1["measurement"]
z = df2["measurement"]

plt.plot(x, y, z)
plt.ylabel("")
plt.title("")
plt.legend("")
plt.show()
```

```python
df2["measurement"].describe()
```

```python
decomposition = sm.tsa.seasonal_decompose(df2["measurement"], period = 365, model='additive')
fig = decomposition.plot()
```

```python
my_stl = STL(endog=df2["measurement"])


stl = my_stl
res = stl.fit()
fig = res.plot()
```
